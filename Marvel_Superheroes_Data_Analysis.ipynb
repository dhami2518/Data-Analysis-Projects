# Marvel Superheroes Character-Data Analysis

```python
#loading the required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the marvel character_stats dataset
marvel=pd.read_csv('/content/sample_data/charcters_stats.csv')
```

```python
# Extracting the first five rows of the datasets
marvel.head()
```

```python
# check the how many rows and columns in this datasets using shape method
marvel.shape
```

```python
# check the frequency of the Alignment in this datasets
marvel['Alignment'].value_counts()
```

```python
marvel[marvel['Alignment']=='good']
```

```python
good=marvel[marvel['Alignment']=='good']
```

```python
good.head()
```

```python
good.sort_values(by=['Speed'],ascending=False)
```

```python
good.sort_values(by=['Speed'],ascending=False).head()
```

```python
good.sort_values(by=['Power'],ascending=False).head()
```

```python
good_max_power_full=good[good['Power']==100]
```

```python
good_max_power_full.shape
```

```python
good_max_power_full
```

```python
good_max_power=good.sort_values(by=['Total'],ascending=False)
```

```python
good_max_power.head()
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(good_max_power['Name'])[0:5],list(good_max_power['Total'])[0:5],color='r')
plt.show()
```
